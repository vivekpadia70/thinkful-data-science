```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
raw_data = pd.read_csv('C:/Users/vivek/Downloads/Thinkful_datasets/epicurious-recipes-with-rating-and-nutrition/epi_r.csv')
```

```python
raw_data.head()
```

```python
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
Y = raw_data['rating']
```

Because of more number of NaN values we dropped nutrition values from the dataframe.

```python
Y.mean()
```

For label we should be trying to maintain ratio of both the values. So, we will be seperating both from the point 4.0.

```python
Y = Y > 4.0
```

```python
Y = Y.replace(True, 1)
```

```python
plt.hist(Y)
```

### PCA with 95% variance

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
pca = PCA(.95)
X = pca.fit_transform(X)
```

```python
svc = SVC(gamma='auto')
svc.fit(X, Y)
```

```python
svc.score(X,Y)
```

```python
cross_val_score(svc, X, Y, cv=5)
```

```python
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
```

### PCA with 30 features

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
pca = PCA(n_components=30)
X = pca.fit_transform(X)
```

Let's check if this data is somehow biased or not. First, we will try by checking if one ingredient is creating any bias.

```python
df = raw_data.drop(['title', 'rating', 'calories', 'protein', 'fat', 'sodium', '#cakeweek', '#wasteless', '22-minute meals', '3-ingredient recipes', '30 days of groceries', 'advance prep required'],1)
```

```python
name = []
times_added = []
```

```python
for x in df.columns:
    name.append(x)
    times_added.append(df[x].sum())
```

```python
d = {'name':name, 'times_added':times_added}
total = pd.DataFrame(data=d)
total.sort_values('times_added', ascending=False)
```

It doesn't show any signs to create bias with ingredients. Next, we would check if most rated recipes have some pattern in nutritions.

```python
df = raw_data[raw_data['rating'] == raw_data['rating'].max()]
print(df)
```

It looks like high calories and high fat foods are more likely to get good ratings but that's just the human nature. So, i don't find any bias in the data.
