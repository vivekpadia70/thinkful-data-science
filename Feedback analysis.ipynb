```python
import sklearn
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_csv('/home/vivek/Downloads/sentiment labelled sentences/imdb_labelled.txt', delimiter='\t', header=None)
```

```python
df.columns = ['review', 'label']
```

```python
df.head()
```

```python
negative_words = ['lose', 'loss', 'kill', 'bad', 'hate', 'fake', 'poor', 'lost', 'low', 'cut', 'worst', 'slow', 'little', 'ridiculous', 'disappoint', 'unhappy']
```

```python
for word in negative_words:
    df[str(word)] = df['review'].str.contains(str(word), case=False)
```

```python
df
```

```python
naive = BernoulliNB()
```

```python
naive.fit(df[negative_words], df['label'])
```

```python
y_pred = naive.predict(df[negative_words])
```

```python
print("Number of mislabeled points out of a total {} points : {}".format(df.shape[0], (df['label'] != y_pred).sum()))
```
