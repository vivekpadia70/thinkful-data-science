```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
sns.pairplot(df, kind='reg', palette='colorblind', hue='species', diag_kind='kde')
```

These graphs show relationship between all columns with different species. Lines in scatter plot show regression and 95% confidence interval with it. These type of graphs are useful for understanding data in initial stages.

```python
sns.boxplot(y=df['sepal_length'])
```

Boxplot is used for understanding characteristics of certain column of dataset. It shows distribution of datapoints and could also identify outliers.

```python
sns.barplot(y=df['sepal_length'])
```

Barplot is used for showing values of different columns. But most of the space is wasted in this graph.

```python
sns.pointplot(df['sepal_length'])
```

Pointplot is gives range with it's mean.

```python
sns.rugplot(df['sepal_length'], color='r')
```

rugplot shows every point with small lines.

```python
plt.figure(figsize=(10,5))
sns.countplot(df['sepal_length'])
```

countplot shows count of each point from a single variable.

```python
sns.kdeplot(df['sepal_length'], df['sepal_width'], shade=True)
```

kdeplot gives an idea about relationship between two variables.

```python
sns.jointplot(df['sepal_length'], df['sepal_width'], kind='reg')
```

jointplot gives much information with scatterplot and histograms on same graph.

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=df, hue='species')
```

lm stands for linear model and it gives idea about relationship between two variables.

```python
sns.violinplot(df['species'], df['sepal_width'])
```

violin plot is same as boxplot but gives more information about variability of points.

```python
sns.barplot(x='species', y='sepal_width', data=df)
```

These are bars from barplot as discussed above.

```python
sns.boxplot(x='species', y='sepal_width', data=df)
```

```python
sns.swarmplot(x='species', y='sepal_width', data=df, split=True)
```

swarmplot is pointed version of violin plot.

```python
sns.stripplot(x='species', y='sepal_width', data=df)
```

stripplot shows single occurences of each point in a variable with points

```python
sns.distplot(df['sepal_width'], rug=True)
```

distplot shows histogram of a variable along with kde
