```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)
```

```python
sample1 = np.random.choice(pop1, 100)
sample2 = np.random.choice(pop2, 100)
```

```python
print(sample1.mean())
print(sample1.std())
print(sample2.mean())
print(sample2.std())
```

```python
plt.figure(figsize=(10, 5))
plt.style.use('ggplot')
plt.subplot(1, 2, 1)
plt.hist(sample1)
plt.subplot(1, 2, 2)
plt.hist(sample2)
plt.show()
```

```python
sample1 = np.random.choice(pop1, 1000)
sample2 = np.random.choice(pop2, 1000)
plt.figure(figsize=(10, 5))
plt.style.use('ggplot')
plt.subplot(1, 2, 1)
plt.hist(sample1)
plt.subplot(1, 2, 2)
plt.hist(sample2)
plt.show()
print(sample1.mean())
print(sample1.std())
print(sample2.mean())
print(sample2.std())
```

```python
sample1 = np.random.choice(pop1, 20)
sample2 = np.random.choice(pop2, 20)
plt.figure(figsize=(10, 5))
plt.style.use('ggplot')
plt.subplot(1, 2, 1)
plt.hist(sample1)
plt.subplot(1, 2, 2)
plt.hist(sample2)
plt.show()
print(sample1.mean())
print(sample1.std())
print(sample2.mean())
print(sample2.std())
```

##### Every value changes between two samples
