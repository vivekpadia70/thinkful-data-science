```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Normal Distribution

```python
np.random.seed(33)
normalVar = np.random.normal(1,2,100)
```

```python
plt.style.use('ggplot')
plt.hist(normalVar)
plt.title('normal distribution')
plt.axvline(normalVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normalVar.mean() - normalVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normalVar.mean() + normalVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

# Poisson Distribution

```python
poissonVar = np.random.poisson(5, 100)
plt.hist(poissonVar)
plt.title('poisson distribution')
plt.axvline(poissonVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poissonVar.mean() - poissonVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poissonVar.mean() + poissonVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

# Binomial Distribution

```python
binomialVar = np.random.binomial(15, 0.5, 100)
plt.hist(binomialVar)
plt.title('binomial distribution')
plt.axvline(binomialVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomialVar.mean() - binomialVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomialVar.mean() + binomialVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

# Exponential Distribution

```python
exponentialVar = np.random.exponential(0.5, 100)
plt.hist(exponentialVar)
plt.title('exponential distribution')
plt.axvline(exponentialVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponentialVar.mean() - exponentialVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponentialVar.mean() + exponentialVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

# Triangular Distribution

```python
triangularVar = np.random.triangular(-2, 0, 8, 100)
plt.hist(triangularVar)
plt.title('triangular distribution')
plt.axvline(triangularVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(triangularVar.mean() - triangularVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(triangularVar.mean() + triangularVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

# Power Distribution

```python
powerVar = np.random.power(10, 100)
plt.hist(powerVar)
plt.title('power distribution')
plt.axvline(powerVar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(powerVar.mean() - powerVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(powerVar.mean() + powerVar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

##### In conclusion we can say that normal distribution and triangular distribution have data centered to center. Poisson and Exponential distribution have data centered to left. Power distribution has data centered to right and for binomial distribution it can't be said.

```python
normalVar1 = np.random.normal(5, 0.5, 100)
normalVar2 = np.random.normal(10, 1, 100)
normalVarTotal = normalVar1 + normalVar2
plt.hist(normalVarTotal)
plt.axvline(normalVarTotal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normalVarTotal.mean() - normalVarTotal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normalVarTotal.mean() + normalVarTotal.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

##### By adding two normal distributions we get distribution which has mean at center but the data is not at center.
