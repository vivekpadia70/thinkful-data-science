```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/vivek/Downloads/LoanStats3d.csv/LoanStats3d.csv',skipinitialspace=True,
    header=1)
```

```python
df.drop(['id', 'member_id'], 1)
```

```python
df.shape
```

```python
categorical = df.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
df.drop(['url', 'sub_grade', 'emp_title', 'desc', 'zip_code', 'addr_state', 'earliest_cr_line', 'revol_util', 'last_pymnt_d', 'last_credit_pull_d', 'hardship_start_date', 'hardship_end_date', 'debt_settlement_flag_date', 'settlement_date'], 1, inplace=True)
```

```python
df
```

```python
df = df[['loan_amnt', 'funded_amnt', 'term', 'int_rate', 'installment', 'grade', 'home_ownership', 'annual_inc', 'verification_status', 'loan_status', 'purpose', 'dti', 'open_acc', 'revol_bal', 'total_acc', 'total_pymnt_inv', 'tot_cur_bal', 'total_il_high_credit_limit']]
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
categorical = df.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
df['int_rate'] = pd.to_numeric(df['int_rate'].str.strip('%'), errors='coerce')
```

```python
X = df.drop('loan_status', 1)
Y = df['loan_status']
X = pd.get_dummies(X)

rfc = ensemble.RandomForestClassifier(n_estimators=10)

cross_val_score(rfc, X, Y, cv=10)
```
