Mall uses website for online availability of services provided by them. Developers suggests that changing the look of homepage of their website could increase the sales of products and could possibly attract more spending customers. So, we need to run an A/B test to check whether new design works or not.

We need following things before rolling out our test.

**Copy of both websites:** So it is clear which are two options in our test.

**Rollout Plan:** We will be showing both the sites to randomly different users. It will be divided in 50/50 proportion. But, for initial purposes we will use 10/90 as if new site breaks then we can rollback easily.
After few weeks, we will change that to 50/50 and start collecting data. This test will run for 2 months and at the end of test apprx 1000+ customer data will be available for further evaluation.
When the test is over and data is in our hands, we can run t-test for both the data and conclude whether our hypothesis is correct or not.

**Success Metric:** Increase in sales and increase in consumers with higher spending score is the success metric of this test.

**Secondary Metrics:** <br> *last_page_by_user*: It will give developers an insight that where users are closing the site.<br>
*ordered_products*: This will be useful for suggesting other products to users based on previously ordered products.
