```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, MeanShift, estimate_bandwidth, SpectralClustering
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
from bokeh.plotting import figure, show

%matplotlib inline
```

we would be using Boston Marathon data from year 2014.

```python
df = pd.read_csv('~/Downloads/boston.csv')
```

```python
df.describe()
```

Here overall has the highest std which means it has highest variance in whole dataset. We should expect most effect of it in our model.

```python
df.head()
```

```python
X = df.drop(['name', 'gender', 'bib', 'ctz', 'city', 'country', 'state'],1)
X = X[X['25k'] != '-']
X = X[X['5k'] != '-']
X = X[X['10k'] != '-']
X = X[X['20k'] != '-']
X = X[X['30k'] != '-']
X = X[X['35k'] != '-']
X = X[X['40k'] != '-']
X = X[X['half'] != '-']
print(X.head())
```

```python
kmeans = KMeans(n_clusters=4, random_state=22)
kmeans.fit(X)
y_pred = kmeans.predict(X)
unique, counts = np.unique(y_pred, return_counts=True)
dict(zip(unique, counts))
sns.scatterplot(X['age'], X['overall'])
```

I have used k-means with 4 clusters. The graph here looks alot conjusted.

```python
sns.scatterplot(X['age'], X['overall'], hue=y_pred)
```

Our k-means explains the difference very well but there will be some confusions when data point will be between two seperate clusters. But our type of dataset makes it hard to make seperate clusters. We will try other graphs after normalizing data.

```python
kmeans = KMeans(n_clusters=5, random_state=22)
kmeans.fit(X)
y_pred = kmeans.predict(X)
unique, counts = np.unique(y_pred, return_counts=True)
dict(zip(unique, counts))
```

## With normalization

```python
# normalize and make pca with 2 features
X = normalize(X)
X_pca = PCA(2).fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=22)
kmeans.fit(X_pca)
y_pred = kmeans.predict(X_pca)
unique, counts = np.unique(y_pred, return_counts=True)
dict(zip(unique, counts))

# plotting scatter plot of both the features of pca
sns.scatterplot(X_pca[:, 0], X_pca[:, 1])
```

After normalization data looks alot more organised and can be used for clustering more efficiently.

```python
sns.scatterplot(X_pca[:, 0], X_pca[:, 1], hue=y_pred)
```

These clusters feels more appealing than what we had seen earlier.

## Mean-Shift

```python
bandwidth = estimate_bandwidth(X_pca, quantile=0.2, n_samples=500)

ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X_pca)

labels = ms.labels_
cluster_centers = ms.cluster_centers_

print(f"Cluster centers: {cluster_centers}")

sns.scatterplot(x_pca[:, 0], X_pca[:, 1], hue=labels)
```

Graph doesn't as good as expected but as we don't have measure to say if model score is good or not so we will go with k-means.

## Spectral Clustering

```python
sc = SpectralClustering(n_clusters=4)

y_pred = sc.fit_predict(X_pca[:3000])

sns.scatterplot(x_pca[:3000, 0], X_pca[:3000, 1], hue=y_pred)
```

We needed to decrease the number of data points for spectral clustering but it's performing better than mean-shift for our particular data.

# Conclusion

It turns out that normalization could make the clustering alot easier for our model along with PCA. But it has some informational loss and if possible we should avoid that for better accuracy. Moreover, If we want to know particular features which affects the model than PCA makes it harder for us as we can't understand each of the features from PCA.

So first we need to make few plots with data without normalization and after we understand the clustering without normalization we could go further and use normalization with PCA for better performance.
